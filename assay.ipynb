```python
'''
投料比为质量比 _ratio
当量比为摩尔比 _equal
'''
feed_ratio = 1.1                            # 投料比
solution_ratio = 4                          # 溶剂投料比   DMF，喹啉
catalyst_ratio = 1.2                        # 催化剂投料比  碳酸钾
specification = 0.995                       # 产品规格
work_day_year = 200                         # 年工作日 天
productive_ability = 50                     # 生产能力 T/a
yields = [0.8,0.83,0.85,0.9]                # 各阶段产率

# NaOH 水解液
NaOH_equal = 2                              # 2 当量的NaOH
NaOH_water_ratio = 4                        # 5倍质量比的水进入

# 活性炭投料比
actvd_cbn_ratio = 0.1                       # 10%

# 盐酸投料比
HCL_equal = 2                               # 2 当量的HCl

# 离心杂质去除比例
centrfg_ratio = 0.2                         # 去除20%杂质

# 洗涤杂质去除比例
wash_deimpty_ratio = 0.8                    # 去除80%的杂质

# 水洗投料比
wash_ratio = 5                      # 5倍质量比的水洗涤

# 氧化亚铜当量比
Cu2O_equal = 0.2              # 催化剂当量  氧化亚铜

# 氮气量
N2_kg = 200

# 淬灭盐酸当量
qench_HCl_equal = 1.5
qench_HCl_water_ratio = 5                 # 5倍质量比的水

# 喹啉蒸出比率
quinoline_steamed = 0.8

# 分液除杂比率
separation_deimpty_ratio = 0.5          # 去除50%的杂质,杂质等量进入水层与有机层

# 乙酸乙酯投料比
ea_ratio = 5

# 氯化钠溶液投料比，氯化钠溶液浓度
NaCl_solution_ratio = 5
NaCl_conc = 0.15
```

```python
indent = '    '
def get_pure_yield_day(pmol,specification,work_day_year,productive_ability):
    kg = productive_ability*1000/work_day_year*specification
    kmol = kg/pmol
    print('日产纯品量为：%.2fkg, %.2fkmol'%(kg,kmol))
    return (kg, kmol)
def get_yield(yields):
    product = 1
    for a_yield in yields:
        product *= a_yield
    product_percent = product*100
    print('总为产率：%.2f%%'%product_percent)
    return product
def snA(n):
    return chr(ord('A')+n-1)+'- '
def sna(n):
    return chr(ord('a')+n-1)+'- '
def od(n):
    return '('+str(n)+') '
```

```python
## functions
import pprint
import math
def GCM(groups, M):
    '''Group Contribution Method
    '''
    cp = 0
    for group in groups:
        if len(group) == 1:
            cp += group[0]
        else:
            cp += group[0]*group[1]
    cp /= M
    print(cp)


def SHCoS(elements, M):
    cp = 0
    for element in elements:
        if len(element) == 1:
            cp += element[0]
        else:
            cp += element[0]*element[1]
    cp /= M
    cp *= 4.1868
    return cp
def get_r(T, Tb, M):
    r = 0
    # 计算沸点时汽化热
    deta_H = Tb/M*(39.8*math.log(Tb, 10)-0.029*Tb)
    # 转化为特定温度下的汽化热
    r = math.pow(T*((1-T)/(1-Tb)), 0.38)*deta_H
    print(r)
```

```python
# DMF 汽化热
get_r(T=60,Tb=153,M=73)
# 盐酸 汽化热
get_r(T=60,Tb=110,M=36.5)
# 喹啉 汽化热
get_r(T=180,Tb=238.05,M=129)
```
